```python
import numpy as np
import pandas as pd
```

```python
sampl = np.random.uniform(low=0.5, high=100, size=(1000,2))
sampl.shape
```

```python
Label=np.random.randint(0,2,size=1000)
Label=Label.reshape(1000,1)
```

```python
sampl=np.hstack((sampl,Label))
```

```python
df=pd.DataFrame(sampl,columns='X Y Label'.split())
```

```python
X7=sampl[:700,:]

X700=pd.DataFrame(X7,columns='X Y Label'.split())
```

```python
b=X700[['X','Y','Label']]

```

```python
Distances=[]
def EuclDist(b,c):
    X_b=b['X']
    Y_b=b['Y']
    X_c=c['X']
    Y_c=c['Y']
    Labels=b['Label']

    Dist=list(np.sqrt((X_b-X_c)**2+(Y_b-Y_c)**2))
    Dist=pd.DataFrame(Dist,columns='EuclideanDist'.split())
    Dist['label']=Labels
    return Dist
```

```python

for j in range(0,300):
    MinDistances=[]
    Y3=sampl[700+j,:]

    X_TestSample=Y3[0]
    Y_TestSample=Y3[1]
    Label_TestSample=Y3[2]
    a=np.hstack((X_TestSample,Y_TestSample))
    c=np.array([a]*700)
    c=pd.DataFrame(c,columns='X Y'.split())
    
    Dist=EuclDist(b,c)
    Dist = Dist.sort_values(by ='EuclideanDist' )
    MinDistances.append(Dist.iloc[0])
    MinDistances.append(Dist.iloc[1])
    MinDistances.append(Dist.iloc[2])
    
    #print('3 Minimum distances :',MinDistances[0],MinDistances[1],MinDistances[2])

    ZeroTerm=0
    OneTerm=0
    for i in range(0,3):
        if MinDistances[i][1]==0.0:
            ZeroTerm = ZeroTerm + 1
            
        else:
            OneTerm = OneTerm + 1
    print('Num of Zeros',ZeroTerm)
    print('Num of Ones',OneTerm)
    if ZeroTerm>OneTerm:
        sampl[700+j,2]=0.0
    else:
        sampl[700+j,2]=1.0

    
    print('Earlier Value',Label_TestSample)
    print('Changed Datapoint',sampl[700+j])
    
    print('-------------------------------------------------------------------------')
```
